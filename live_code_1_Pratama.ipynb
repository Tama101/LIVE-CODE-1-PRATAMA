# Soal 1 :
Buatlah dua fungsi eksplisit masing-masing memuat perhitungan BMR laki-laki dan BMR Perempuan yang nantinya akan digunakan untuk perhitungan

```python
# BMR laki - laki :
# bmr = 66 + (13.7 *BB) + (5 * TB) - (6.8 * U)

# BMR Perempuan :
# bmr = 655 + (9.6 * BB) + (1.8 * TB) - (4.7 * U)

# bb = weight
# tb = height
# u = age
# sex = gender

def bmr_laki(bb, tb, u) :
    bmr = 66 + (13.7 *bb) + (5 * tb) - (6.8 * u)
    return bmr

def bmr_perempuan(bb, tb, u):
    bmr = 655 + (9.6 * bb) + (1.8 * tb) - (4.7 * u)
    return bmr
```

# Soal 2 :
Buat kolom baru dengan nama kolom BMR berisikan hasil perhitungan kebutuhan energi dasar (BMR) harian sesuai dengan jenis kelamin masing-masing pelanggan. (Hint: Gunakan for loop dan if statement untuk mengerjakan perintah nomor b, kamu bisa menggunakan list untuk menyimpan perhitungan sementara, setelahnya dapat dimasukkan ke kolom BMR).

```python
list_bmr =[]

for x in range(len(df)) :
    
    u = df.iloc[x, 0]
    bb = df.iloc[x, 3]
    tb = df.iloc[x, 2]
    sex = df.iloc[x, 1]
    
    if sex == 'Male' :
        bmr = 66 + (13.7 *bb) + (5 * tb) - (6.8 * u)
        
    else:
        sex == 'Female'
        bmr = 655 + (9.6 * bb) + (1.8 * tb) - (4.7 * u)
    list_bmr.append(bmr)
    
df['BMR'] = list_bmr
df
```

# Soal 3
Hitunglah rata-rata berat badan untuk Jenis kelamin laki-laki dengan umur lebih dari sama dengan 24 tahun hingga kurang dari sama dengan 35 tahun.

```python
df[(df['Gender'] == 'Male') &
    (df['Age'] >= 24) &
    (df['Age'] <= 35)].groupby('Gender')['Weight (kg)'].mean().round(1)
```
